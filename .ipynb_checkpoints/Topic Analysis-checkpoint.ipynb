### LSA

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD

def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print('\nTopic ', ix)
        else:
            print('\nTopic ', topic_names[ix])
        print(', '.join([feature_names[i]
                         for i in topic.argsort()[:-no_top_words -1:-1]]))
```

```python
def get_topic_lsa(obj,num_of_topic,thing_in_topic):
    vectorizer = CountVectorizer(stop_words='english',ngram_range=(1, 2))
    doc_word = vectorizer.fit_transform(obj.corpus)
    doc_word.shape

    lsa = TruncatedSVD(num_of_topic)
    doc_topic = lsa.fit_transform(doc_word)
    lsa.explained_variance_ratio_

    topic_world = pd.DataFrame(lsa.components_.round(3), index = list(range(0, num_of_topic)), columns= vectorizer.get_feature_names())

    return display_topics(lsa, vectorizer.get_feature_names(), thing_in_topic)
```

### NMF

```python
from sklearn.decomposition import NMF
```

```python
def get_topic_nmf(obj,num_of_topic,thing_in_topic):
    vectorizer = CountVectorizer(stop_words='english',ngram_range=(1, 2))
    doc_word = vectorizer.fit_transform(obj.corpus)
    nmf_model = NMF(num_of_topic)
    doc_topic = nmf_model.fit_transform(doc_word)

    topic_word = pd.DataFrame(nmf_model.components_.round(3),index =  list(range(0, num_of_topic)), columns = vectorizer.get_feature_names())

    display_topics(nmf_model, vectorizer.get_feature_names(), thing_in_topic)
```
